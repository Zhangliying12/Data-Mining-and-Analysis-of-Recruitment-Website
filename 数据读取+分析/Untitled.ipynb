```python
import pymysql
from sqlalchemy import create_engine
import pandas as pd
import numpy as np
import matplotlib.pylab as pyl
conn=create_engine('mysql+pymysql://root:root@localhost:3306/job_info',encoding='utf8')
def read_from_db(table_name):
    sql = "select * from " + table_name
    df = pd.read_sql(sql,conn)
    return df
attr = [
    'position',
    'salary_min',
    'salary_max',
    'experience',
    'degree',
    'work_place',
    'company_name',
    'company_type',
    'company_finance',
    'company_scale',
    'bonus',
    'url'
]
# 对每一类型的职位进行数据清洗：
def clean_data(df):
    # 去掉重复项
    df = df.drop_duplicates()
    # 去掉含有空值和“”的项
    df = df.dropna()
    for att in attr:
        df = df[~df[att].isin([""])]
    # 去掉work_place值异常的项
    df = df[~df['work_place'].str.contains("{{item.city}}")]
    # 去掉异常值
    df = df[~df['salary_max'].isin([1])]
    return df

data=read_from_db('project_management_copy')
data2=clean_data(data)

pd.io.sql.to_sql(data2,'project_management_copy', conn, if_exists='replace')
```
